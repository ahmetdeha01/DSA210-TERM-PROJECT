## 2. Data Enrichment

The original dataset contains only 50 samples. To improve ML model performance, two enrichment steps are applied:

**Feature Engineering:** 6 new physically meaningful features are derived:
- `heat_ratio` = nozzle_temperature / bed_temperature
- `thermal_delta` = nozzle_temperature - bed_temperature
- `volumetric_flow` = layer_height x print_speed
- `infill_wall_ratio` = infill_density / wall_thickness
- `material_pla` — binary encoding of material type
- `pattern_honeycomb` — binary encoding of infill pattern

**Gaussian Noise Augmentation:** 4 copies of the original data are created with 3% Gaussian noise, expanding the dataset from 50 to 250 samples.

## 3. ML Analysis

Four regression models are trained and evaluated for each of the three output variables:
- **roughness** (surface roughness in µm)
- **tensile_strength** (ultimate tensile strength in MPa)
- **elongation** (elongation at break in %)

**Train/Validation/Test Split:** 60% / 20% / 20%

**Hyperparameter Tuning:** GridSearchCV with 5-fold cross-validation on the training set

**Evaluation Metrics:** RMSE, MAE, R²

**Models:**
- Linear Regression (baseline)
- Random Forest
- XGBoost
- SVR (Support Vector Regression)

```python
# Run ML analysis with GridSearchCV hyperparameter tuning
exec(open('ml_analysis_v2.py').read())
```

## 4. Results Summary

After GridSearchCV tuning, SVR with RBF kernel consistently achieved the best results:

| Target | Best Model | Test R² |
|--------|-----------|--------|
| roughness | SVR | 0.993 |
| tensile_strength | SVR | 0.999 |
| elongation | SVR | 0.997 |

Key findings:
- `layer_height` is the most important feature for roughness prediction (importance = 0.52)
- `wall_thickness` and `infill_density` dominate tensile strength prediction
- Linear Regression struggled with tensile_strength (R²=0.68), confirming non-linear relationships
- Hyperparameter tuning significantly improved SVR performance across all targets
